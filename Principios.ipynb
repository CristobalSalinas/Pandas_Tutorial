```python
import pandas as pd
```

### Crear un Dataframe

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
            ],
           "Age": [22, 35, 58],
           "Sex": ["male", "male", "female"],
      }
)

df
```

### Seleccionando una sola columna de un Dataframe se le llama 'Series'

```python
df['Age']
```

### Podemos crear una "Serie" aparte

```python
edades = pd.Series([23,45,32],name="Edades")
edades
```

### Formulas con Dataframe o Series

```python
df['Age'].max() #Ver cual es la edad mas grande
```

```python
edades.max()
```

```python
df.describe()
```
