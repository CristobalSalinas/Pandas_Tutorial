```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
titanic
```

# Sacar el promedio

```python
titanic["Age"].mean()
```

# Sacar la media

```python
titanic[['Age','Fare']].median()
```

# Describir con las 2 columnas

```python
titanic[['Age','Fare']].describe()
```

# Sacar promedio de hombres vs mujeres

```python
titanic[['Sex','Age']]
```

```python
titanic[['Sex','Age']].groupby('Sex').mean()
```

# De otra manera

```python
titanic.groupby('Sex').mean()
```

```python
titanic.groupby('Sex')['Age'].mean()
```

# Promedio de 'Fare' agrupados por sex y pclass

```python
titanic.groupby(['Sex','Pclass']).mean()
```

# Contar las cantidades de cierta opcion

```python
titanic['Pclass'].value_counts()
```

```python
# De otra manera

titanic.groupby('Pclass')['Pclass'].count()
```
