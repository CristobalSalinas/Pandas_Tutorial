### Puedo seleccionar un dataframe de un dataframe

```python
test = titanic[['Age','Sex']]
test
```

```python
print(test.shape)
```

### Filtrar dataframe por valor especifico

```python
arriba_35 = titanic[titanic['Age'] > 35]

arriba_35
```

```python
arriba_35.shape
```

### alguna condicion especifica

```python
class_23 = titanic[titanic['Pclass'].isin([2,3])]
class_23
```

```python
class_23 = titanic[(titanic['Pclass'] == 2) | (titanic['Pclass'] == 3)]
class_23
```

### Ver pasajeros los cuales su edad si exista

```python
age_no_na = titanic[titanic['Age'].notna()]
age_no_na
```

### Seleccionar rows y columnas especificas

```python
# Quiero obtener los nombres de los pasajeros mas viejos que 35
olders = titanic.loc[titanic['Age']>35,['Name','Sex']]
olders
```

### Obteniendo rows desde el 10 hasta el 25 y columnas desde el 3 al 5

```python
# Imprimiendo primero como se ve normalmente
titanic
```

```python
# ahora como queremos
titanic.iloc[9:25,2:5]
```

```python
titanic.iloc[0:3,3] = "Anonimo"
titanic.head()
```

### Entonces para recordar
#### * Para seleccionar data dentro del dataframe lo hacemos con []
#### * dentro de estas podemos usar un simple nombre de columna o row, una lista, condiciones etc
#### * loc para obtener las rows y columas
#### * iloc para posiciones de las tablas
#### * de iloc/loc podemos asignarles nuevos valores
